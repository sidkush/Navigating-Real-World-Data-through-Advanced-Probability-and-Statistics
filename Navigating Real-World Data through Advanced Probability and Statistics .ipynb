# Simulating Continuous Random Variables

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,3))

# Set the parameters of the gamma distribution
shape_parameter = 2.5  # Shape parameter (a)
scale_parameter = 1.0  # Scale parameter (scale)

# Generate random samples from a gamma distribution
random_samples = np.random.gamma(shape_parameter, scale_parameter, 1000)

# Plot the histogram of the random samples
plt.hist(random_samples, bins=30, density=True, alpha=0.7, color='blue')

# Plot the probability density function (PDF) of the gamma distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = (1/(scale_parameter**shape_parameter) * x**(shape_parameter-1) * np.exp(-x/scale_parameter))
plt.plot(x, pdf, 'k-', linewidth=2, color='red')

# Add labels and a title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Gamma Distribution')

# Show the plot
plt.show()
```

We can visualize the random gamma distribution generated through numpy libraries. Clearly, the data looks left skewed. Now, performing statistical analysis on the data to understand it better.

## Statistical Analysis

```python
# Calculate statistical measures

mean = np.mean(random_samples)

variance = np.var(random_samples)
std_deviation = np.std(random_samples)

first_quantile = np.percentile(random_samples, 25)
third_quantile = np.percentile(random_samples, 75)

# Mode for gamma distribution is (a - 1) * scale
mode = shape_parameter * scale_parameter - 1   

skewness = moment(random_samples, moment=3) / (np.std(random_samples) ** 3)
kurtosis = moment(random_samples, moment=4) / (np.std(random_samples) ** 4) - 3

# Print the calculated measures
print(f"Mean: {mean}\n")

print(f"Variance: {variance}\n")
print(f"Standard Deviation: {std_deviation}\n")

print(f"First Quantile (Q1): {first_quantile}\n")
print(f"Third Quantile (Q3): {third_quantile}\n")

print(f"Mode: {mode}\n")

print(f"Skewness: {skewness}\n")
print(f"Kurtosis: {kurtosis}\n")
```

This data gives us an idea about the distribution of the data. Mean is equal to 2.48 and standard deviation being 1.56. The major chunk of data lies between 1.32 to 3.32, i.e. the Inter-Quantile Range (IQR).

## Central Limit Theorem Verification

The Central Limit Theorem (CLT) states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables approaches a normal distribution, regardless of the original distribution. Let's verify the CLT by taking random samples from the simulated gamma distribution and examining the distribution of the sample means.

```python
# Number of samples to take and size of each sample
num_samples = 1000
sample_size = 30

# Calculate sample means
sample_means = [np.mean(np.random.choice(random_samples, sample_size)) for _ in range(num_samples)]

# Plot the distribution of sample means
plt.figure(figsize=(10, 3))

# Plot the histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='magenta', label='Sample Means')

# Plot the theoretical normal distribution based on CLT
mean_of_means = np.mean(sample_means)
std_dev_of_means = np.std(sample_means, ddof=1)  # ddof=1 for sample standard deviation

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_of_means, std_dev_of_means)

plt.plot(x, p, 'k', linewidth=2, color='green', label='Normal Distribution (CLT)')

# Add labels and a title
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('Distribution of Sample Means and Normal Distribution (CLT)')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

Here:

- Generated 1000 random samples, each with a size of 30, from the simulated gamma distribution.
- Calculated the mean of each sample and store them in the sample_means list.
- Plotted a histogram of the sample means and overlay a normal distribution curve with the same mean and standard deviation calculated from the sample means.

If the sample size is sufficiently large, the distribution of sample means should approximate a normal distribution, as predicted by the Central Limit Theorem. The larger the sample size, the better the approximation.

# Markov Chains

Markov Chains are stochastic processes that involve a sequence of events, where the probability of transitioning to any particular state depends solely on the current state and time elapsed. These chains have the Markov property, meaning that the future state of the system depends only on the current state and not on how the system arrived at its current state.

The key components of a Markov Chain are:

- States: A set of possible values that a random variable (system) can take.
- Transition Probabilities: The probabilities of moving from one state to another in the next step.

## Transition Matrix Simulation

Creating a simulation of a simple Markov chain with a specified transition matrix. We'll model transitions between states and calculate the state probabilities after a certain number of steps.

Here's an example with a 3-state Markov chain:

```python
# Define the transition matrix
transition_matrix = np.array([
    [0.7, 0.2, 0.1],
    [0.3, 0.6, 0.1],
    [0.2, 0.3, 0.5]
])

# Define the initial state probabilities
initial_state_probabilities = np.array([0.3, 0.4, 0.3])

# Number of steps in the simulation
num_steps = 10

# Simulate the Markov chain
current_state = np.random.choice(range(len(initial_state_probabilities)), p=initial_state_probabilities)
for step in range(num_steps):
    print(f"Step {step + 1}: State {current_state + 1}")
    current_state = np.random.choice(range(len(transition_matrix)), p=transition_matrix[current_state])

# Calculate state probabilities after the specified number of steps
# initial_state_prob (I) * transition_matrix (T)^(num_steps[10])
# I.T^10
final_state_probabilities = np.dot(initial_state_probabilities, np.linalg.matrix_power(transition_matrix, num_steps))

# Print the final state probabilities
print(f"\nState Probabilities after {num_steps} steps:")
for i, prob in enumerate(final_state_probabilities):
    print(f"State {i + 1}: {prob:.4f}")
```

Here:

- We define a transition matrix **'transition_matrix'** representing the probabilities of transitioning between different states.
- We define the initial state probabilities **'initial_state_probabilities'**.
- We specify the number of steps in the simulation with **'num_steps'**.
- We simulate the Markov chain by randomly selecting the next state based on the transition probabilities.
- We calculate the state probabilities after the specified number of steps using matrix exponentiation.

## Recurrent Events

 Modeling recurrent events using a Markov chain is a common approach in simulating systems with dynamic and sequential processes. Let's consider a simple example of a queueing system with customer arrivals, service times, and departures. In this scenario, we can define states representing the number of customers in the system, and transitions representing events like arrivals, service completion, and departures.
   
 

```python
# Parameters for the Markov chain
arrival_rate = 2.0  # customers per time unit
service_rate = 1.5  # customers per time unit
max_customers = 5  # maximum number of customers in the system
num_steps = 100  # number of time steps

# Simulate the Markov chain
current_state = 0
state_counts = np.zeros(max_customers + 1)

for step in range(num_steps):
    
    # Generate random events
    arrival_event = np.random.poisson(arrival_rate)
    departure_event = np.random.poisson(service_rate)

    # Update state based on events
    current_state = max(0, current_state + arrival_event - departure_event)
    current_state = min(max_customers, current_state)

    # Record the state
    state_counts[current_state] += 1
    
# Plot the results
plt.figure(figsize=(10,2))
plt.bar(range(max_customers + 1), state_counts / num_steps, color='skyblue')
plt.xlabel('Number of Customers in System')
plt.ylabel('Probability')
plt.title('Recurrent Events Simulation in Queueing System')
plt.show()
```

Here:

- We have initialized value of **arrival_rate**, **service_rate** and **max_customers** as 2, 1.5 and 5 respectivey. They are being random generated using poisson distribution i.e. our transition matrix.
- We are updating the current state by using **current_state** + **arrival_rate** - **service_rate** i.e. current state describes the current number of customers at the place, arrival_rate gives us number of customers arriving at a particular time and service rate describes the number of customers who have been serviced or departing the place.
- We are simulating for 100 time steps.
- The histogram represents the probability of occuring each state throughout the transition and it will change everytime as the transition matrix itself is dynamic and changes with the poisson distribution.

## Ergodicity

Simulating a Markov chain and comparing the time-averaged behavior to the steady-state probabilities can provide insights into the convergence of the Markov chain. In a steady state, the time-averaged behavior should converge to the steady-state probabilities.

Creating an example of a 3-state Markov chain and compare the time-averaged behavior with the steady-state probabilities:

```python
# Number of time steps
num_steps = 1000

# Simulate the Markov chain
current_state = np.random.choice(range(len(transition_matrix)))
state_counts = np.zeros(len(transition_matrix))

for step in range(num_steps):
    state_counts[current_state] += 1
    current_state = np.random.choice(range(len(transition_matrix)), p=transition_matrix[current_state])

# Calculate time-averaged behavior
time_averaged_behavior = state_counts / num_steps

# Calculate steady-state probabilities
steady_state_probabilities = np.linalg.eig(transition_matrix.T)[1][:, 0]
steady_state_probabilities /= np.sum(steady_state_probabilities)

# Print results
print("Time-Averaged Behavior:")
print(time_averaged_behavior)
print("\nSteady-State Probabilities:")
print(steady_state_probabilities)
```

Here:

- We define a 3-state Markov chain with a transition matrix.
- We simulate the Markov chain for a specified number of time steps (**'num_steps'**).
- We track the number of visits to each state to calculate the time-averaged behavior.
- We calculate the steady-state probabilities using the eigenvalues of the transpose of the transition matrix.

The time-averaged behavior and steady-state probabilities are then printed for comparison. In an ideal scenario, as the number of time steps increases, the time-averaged behavior should converge to the steady-state probabilities. Here it is approximately equal to the steady state. <br><br>
Here, this matrix does not follow ergodicity.

## Sensitivity Analysis

Sensitivity analysis involves assessing how small changes in parameters affect the behavior of a system. In the context of Markov chains, this could involve simulating the chain with different transition probabilities or initial conditions. Let's perform sensitivity analysis on a simple 3-state Markov chain by varying transition probabilities:

```python
def simulate_markov_chain(transition_matrix, initial_state, num_steps):
    current_state = initial_state
    state_counts = np.zeros(len(transition_matrix))

    for step in range(num_steps):
        state_counts[current_state] += 1
        current_state = np.random.choice(range(len(transition_matrix)), p=transition_matrix[current_state])

    return state_counts / num_steps

def sensitivity_analysis(base_transition_matrix, perturbation_factor, num_steps):
    num_states = len(base_transition_matrix)
    num_simulations = 100

    # Perform sensitivity analysis by perturbing transition probabilities
    results = []

    for _ in range(num_simulations):
        perturbed_matrix = base_transition_matrix + np.random.uniform(-perturbation_factor, perturbation_factor, size=(num_states, num_states))
        
        # Ensure probabilities stay between 0 and 1
        perturbed_matrix = np.clip(perturbed_matrix, 0, 1)
        
        # Normalize rows to sum to 1
        perturbed_matrix /= perturbed_matrix.sum(axis=1, keepdims=True)

        result = simulate_markov_chain(perturbed_matrix, initial_state=0, num_steps=num_steps)
        results.append(result)

    return results

plt.figure(figsize=(25,28))

# Base transition matrix
base_transition_matrix = np.array([
    [0.7, 0.2, 0.1],
    [0.3, 0.6, 0.1],
    [0.2, 0.3, 0.5]
])

# Parameters for sensitivity analysis
perturbation_factor = 0.1
num_steps = 1000

# Perform sensitivity analysis
results = sensitivity_analysis(base_transition_matrix, perturbation_factor, num_steps)

# Plot results
for i, result in enumerate(results):
    plt.plot(result, label=f'Simulation {i + 1}')

# Plot the base case
base_result = simulate_markov_chain(base_transition_matrix, initial_state=0, num_steps=num_steps)
plt.plot(base_result, label='Base Case', linewidth=6.5, linestyle='--', color='black')


# Add labels and legend
plt.xlabel('State', fontsize=36)
plt.ylabel('Probability', fontsize=36)
plt.title('Sensitivity Analysis of Markov Chain',fontsize=46)
plt.xticks(fontsize=26)
plt.yticks(fontsize=26)
plt.legend()
plt.show()
```

Here:

- **'simulate_markov_chain'** function simulates a Markov chain given a transition matrix, initial state, and number of steps.
- **'sensitivity_analysis'** function performs sensitivity analysis by perturbing the transition probabilities within a specified range and simulating the chain multiple times.
- We use a base transition matrix and perturb its values to observe the sensitivity of the system.

The plot shows the results of multiple simulations, each with a perturbed transition matrix. You can observe how small changes in transition probabilities affect the system's behavior.

# Variance Reduction Techniques

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the target distribution (e.g., normal distribution)
mu = 0
sigma = 1
target_distribution = lambda x: 1/(np.sqrt(2*np.pi)*sigma) * np.exp(-(x-mu)**2/(2*sigma**2))

# Define the proposal distribution (e.g., a normal distribution with a different mean)
proposal_distribution = lambda x: 1/(np.sqrt(2*np.pi)*2) * np.exp(-(x-2)**2/8)

# Generate samples from the proposal distribution
num_samples = 10000
samples = np.random.gamma(2.5, 1.0, num_samples)

# Compute importance weights
weights = target_distribution(samples) / proposal_distribution(samples)

# Estimate the expectation using importance sampling
estimated_expectation = np.mean(weights)
estimated_var = np.var(weights)

# Plotting
plt.figure(figsize=(10, 3))
x = np.linspace(-5, 10, 100)
plt.plot(x, target_distribution(x), label='Target Distribution')
plt.plot(x, proposal_distribution(x), label='Proposal Distribution')
plt.scatter(samples, np.zeros_like(samples), color='red', marker='|', label='Samples')
plt.legend()
plt.title('Importance Sampling')
plt.show()

print(f'Estimated Expectation: {estimated_expectation}')
print(f'Estimated Variance: {estimated_var}')
```

In the above graph, we can observe that the data is being distrinuted uniformally in the orange resamples curve reducing the variance in the data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the target function (e.g., x^2)
target_function = lambda x: x**2

# Define the control variate (e.g., x)
control_variate = lambda x: x

# Estimate the mean of the target function using control variates
estimated_mean = np.mean(target_function(samples) - control_variate(samples)) + np.mean(control_variate(samples))
estimated_var = np.var(target_function(samples) - control_variate(samples)) + np.var(control_variate(samples))

# Plotting
plt.figure(figsize=(10, 3))
x = np.linspace(0, 1, 100)
plt.plot(x, target_function(x), label='Target Function')
plt.plot(x, control_variate(x), label='Control Variate', color = 'yellow')
plt.scatter(samples, target_function(samples), color='red', marker='.', label='Samples')
plt.legend()
plt.title('Control Variates')
plt.show()

print(f'Estimated Mean: {estimated_mean}')
print(f'Estimated Variance: {estimated_var}')
```

Using a control variate we were not able to reduce the variance in this case.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the target function (e.g., x^2)
target_function = lambda x: x**2

# Generate positive and negative pairs of antithetic samples
num_samples = 1000
positive_samples = samples
negative_samples = 1 - positive_samples

# Combine positive and negative samples
samples = np.concatenate([positive_samples, negative_samples])

# Estimate the mean of the target function using antithetic variates
estimated_mean = 0.5 * (np.mean(target_function(positive_samples)) + np.mean(target_function(negative_samples)))
estimated_var = 0.25 * (np.var(target_function(positive_samples)) + np.var(target_function(negative_samples)))

# Plotting
plt.figure(figsize=(10, 3))
x = np.linspace(0, 1, 100)
plt.plot(x, target_function(x), label='Target Function')
plt.scatter(positive_samples, target_function(positive_samples), color='red', marker='.', label='Positive Samples')
plt.scatter(negative_samples, target_function(negative_samples), color='blue', marker='.', label='Negative Samples')
plt.legend()
plt.title('Antithetic Variates')
plt.show()

print(f'Estimated Mean: {estimated_mean}')
print(f'Estimated Variance: {estimated_var}')
```

Antithetic introduces the negative samples to balance out the data. We have reduced the variance in comparison to control variate.

- MCMC Sampling
- Log Transformation
- Importance Sampling (Uniform/Gamma)

```python
import numpy as np
import matplotlib.pyplot as plt
import emcee

# Function to simulate a random process (replace with your own simulation function)
def simulate_random_process(size=10000):
    return np.random.gamma(2.5, 1.0, size)

# Number of simulations
num_simulations = 10000

# Simulate the original process
original_simulations = [simulate_random_process() for _ in range(num_simulations)]
print("Mean (Original Sample):",np.mean(original_simulations))
print("Stddev (Original Sample):",np.std(original_simulations))
print("Variance (Original Sample):",np.var(original_simulations),'\n')

# Simulate the antithetic process
antithetic_simulations = [-(simulate_random_process()) for _ in range(num_simulations)]
print("Mean (Antithetic Sample):",np.mean(antithetic_simulations))
print("Stddev (Antithetic Sample):",np.std(antithetic_simulations))
print("Variance (Antithetic Sample):",np.var(antithetic_simulations),'\n')

# MCMC Simulation
def log_likelihood(theta, data):
    return -0.5 * np.sum((data - theta[0])**2)

def log_prior(theta):
    return 0.0  # Flat prior

def log_posterior(theta, data):
    return log_prior(theta) + log_likelihood(theta, data)

# Set up MCMC
n_walkers = 10
n_steps = 100
n_dim = 1
initial_theta = [1.0]

sampler = emcee.EnsembleSampler(n_walkers, n_dim, log_posterior, args=(original_simulations[0],))

# Run MCMC
pos, _, _ = sampler.run_mcmc(np.random.randn(n_walkers, n_dim) + initial_theta, n_steps)

print("Mean (MCMC Sample):",np.mean(sampler.flatchain))
print("Stddev (MCMC Sample):",np.std(sampler.flatchain))
print("Variance (MCMC Sample):",np.var(sampler.flatchain),'\n')

# Log Transform Simulation
log_transformed_simulations = [np.log(np.abs(simulation) + 1) for simulation in original_simulations]
print("Mean (Log Transformed Sample):",np.mean(log_transformed_simulations))
print("Stddev (Log Transformed Sample):",np.std(log_transformed_simulations))
print("Variance (Log Transformed Sample):",np.var(log_transformed_simulations),'\n')

# Define the target distribution (e.g., normal distribution)
mu = 0
sigma = 1
target_distribution = lambda x: 1/(np.sqrt(2*np.pi)*sigma) * np.exp(-(x-mu)**2/(2*sigma**2))

# Define the proposal distribution (e.g., a normal distribution with a different mean)
proposal_distribution = lambda x: 1/(np.sqrt(2*np.pi)*2) * np.exp(-(x-2)**2/8)

# Compute importance weights
weights = np.array([target_distribution(simulation) for simulation in original_simulations]) / np.array([proposal_distribution(simulation) for simulation in original_simulations])
print("Mean (Importance Sampling):",np.mean(weights))
print("Stddev (Importance Sampling):",np.std(weights))
print("Variance (Importance Sampling):",np.var(weights),'\n')
```

Clearly, in our scenario of gamma distribution MCMC sampling method and Log tranformation are significantly reducing the variance of the distribution.

```python
import numpy as np

def simulate_card_draws(num_simulations=100000):
    deck = np.arange(1, 53)  # Standard deck of 52 cards
    num_cards_in_hand = 5  # Number of cards in a hand

    pairs = 0
    two_pairs = 0
    three_of_a_kind = 0

    for _ in range(num_simulations):
        np.random.shuffle(deck)
        hand = deck[:num_cards_in_hand]

        # Check for pairs, two pairs, and three of a kind
        unique_values, counts = np.unique(hand % 13, return_counts=True)

        if 2 in counts:
            pairs += 1

        if np.sum(counts == 2) == 2:
            two_pairs += 1

        if 3 in counts:
            three_of_a_kind += 1

    # Calculate probabilities
    prob_pair = pairs / num_simulations
    prob_two_pairs = two_pairs / num_simulations
    prob_three_of_a_kind = three_of_a_kind / num_simulations

    return prob_pair, prob_two_pairs, prob_three_of_a_kind

# Run simulations
prob_pair, prob_two_pairs, prob_three_of_a_kind = simulate_card_draws()

# Display results
print(f"Probability of getting a pair: {prob_pair:.4f}")
print(f"Probability of getting two pairs: {prob_two_pairs:.4f}")
print(f"Probability of getting three of a kind: {prob_three_of_a_kind:.4f}")
```

Creating the function to change the cast types to optimmize the memory usage and improve the dataset processing speed.

```python
def cast_data_type(df):
    """
    Optimize the memory usage of a Pandas DataFrame by casting data types.

    Parameters:
    - df (pd.DataFrame): The DataFrame to be optimized.

    Returns:
    - pd.DataFrame: The optimized DataFrame.
    """

    # Get initial memory usage
    start_mem = df.memory_usage(index=True, deep=True).sum() / 1024**2
    start_mem_GB = df.memory_usage(index=True, deep=True).sum() / 1024**3
    print(f'Initial memory usage of dataframe is {start_mem:.2f} MB/{start_mem_GB:.2f} GB')

    # Iterate through columns to optimize data types
    for col in df.columns:
        col_type = df[col].dtype

        # Check if the column is not of object type
        if col_type != object:
            col_min = df[col].min()
            col_max = df[col].max()

            # Optimize integer data types
            if str(col_type)[:3] == 'int':
                if col_min > np.iinfo(np.int8).min and col_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif col_min > np.iinfo(np.int16).min and col_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif col_min > np.iinfo(np.int32).min and col_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif col_min > np.iinfo(np.int64).min and col_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            else:
                # Optimize float data types
                if col_min > np.finfo(np.float16).min and col_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif col_min > np.finfo(np.float32).min and col_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        else:
            # Convert object type to category
            df[col] = df[col].astype('category')

    # Get memory usage after optimization
    end_mem = df.memory_usage(index=True, deep=True).sum() / 1024**2
    end_mem_GB = df.memory_usage(index=True, deep=True).sum() / 1024**3
    reduction = 100 * (start_mem - end_mem) / start_mem

    # Print optimization results
    print(f'Memory usage after optimization is: {end_mem:.2f} MB/{end_mem_GB:.2f} GB')
    print(f'Decreased by {reduction:.1f}%')

    return df
```

Creating a funciton to preprocess the data i.e. merging, dropping irrelvant features, optimmize memory usage , encooding the data and truncating the data for faster processing.

```python
def preprocess_data(data1, data2, value):
    """
    Preprocess the input data by merging, dropping columns, optimizing memory usage, and applying label encoding.

    Parameters:
    - data1 (pd.DataFrame): The first DataFrame.
    - data2 (pd.DataFrame): The second DataFrame.
    - value (int): The value to truncate the DataFrame after label encoding.

    Returns:
    - pd.DataFrame: The preprocessed and encoded DataFrame.
    """

    # Merge data1 and data2 on 'TransactionID'
    df = data1.merge(data2, how='left', on='TransactionID')

    # Drop unnecessary columns
    df.drop('TransactionDT', axis=1, inplace=True)

    # Optimize memory usage and cast data types
    df = cast_data_type(df)

    # Apply label encoding to all columns
    df = df.apply(LabelEncoder().fit_transform)

    # Truncate the DataFrame after label encoding
    df = df.truncate(after=value)

    return df
```

## Normality test for Transaction Amount

[3] Normality tests are statistical tests used to assess whether a sample comes from a population with a normal distribution. Two common normality tests are the Shapiro-Wilk test and the Kolmogorov-Smirnov test. Here's an example of how one can perform these tests in Python:

```python
import numpy as np
from scipy.stats import shapiro, kstest
import matplotlib.pyplot as plt

# Simulate data from a normal distribution
simulated_data = train_data['TransactionAmt']

# Visualize the data distribution
plt.hist(simulated_data, bins=30, density=True, alpha=0.5, color='blue')
plt.title("Simulated Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Shapiro-Wilk Test
stat_sw, p_value_sw = shapiro(simulated_data)
print(f"Shapiro-Wilk Test - p-value: {p_value_sw}")

# Kolmogorov-Smirnov Test
stat_ks, p_value_ks = kstest(simulated_data, 'norm')
print(f"Kolmogorov-Smirnov Test - p-value: {p_value_ks}")
```

Clealry, the p-value is 0. Hence, we can reject the null hypothesis and can say that stastically the distribution of transaction amount is not following a normal distribution or not satisfying the assumptions before applying the machine leanring models.
